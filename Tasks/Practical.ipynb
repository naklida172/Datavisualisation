# 04.02 Data Cleaning and Wrangling Workflow

This is the heart of the project.

Because the task is explicitly called a **Data Wrangling Project**, students must show their cleaning process clearly—not just the final cleaned output.

## Core Cleaning Expectations

Students should demonstrate, when applicable:

1. Missing value detection and handling  
2. Column name standardization  
3. Data type conversion  
4. Duplicate detection and removal  
5. Inconsistent text cleaning  
6. Outlier identification and treatment  
7. Validation of the cleaned dataset

```python
# Example Kyrgyzstan-style synthetic dataset (regional socio-economic indicators)
regions = ["Bishkek", "Osh", "Chuy", "Issyk-Kul", "Naryn", "Talas", "Jalal-Abad", "Batken"]
years = np.arange(2019, 2025)

rows = []
for y in years:
    for r in regions:
        population = np.random.randint(150000, 1300000)
        unemployment = np.random.uniform(3.5, 11.5)
        avg_income = np.random.uniform(12000, 45000)
        school_enrollment = np.random.uniform(72, 99)
        hospital_beds = np.random.uniform(1.8, 7.5)
        rows.append([r, y, population, unemployment, avg_income, school_enrollment, hospital_beds])

kg_df = pd.DataFrame(rows, columns=[
    "region", "year", "population", "unemployment_rate", "avg_monthly_income",
    "school_enrollment_rate", "hospital_beds_per_1000"
])

# Intentionally add some realistic data quality issues
kg_df.loc[3, "avg_monthly_income"] = np.nan
kg_df.loc[10, "region"] = "chuy"
kg_df = pd.concat([kg_df, kg_df.iloc[[5]]], ignore_index=True)

kg_df.head()
```

```python
kg_df.info()
```

```python
# 1. Standardize column names
kg_df.columns = [c.strip().lower().replace(" ", "_") for c in kg_df.columns]
kg_df.columns
```

```python
# 2. Fix inconsistent text values
kg_df["region"] = kg_df["region"].replace({"chuy": "Chuy"})
kg_df["region"].unique()
```

```python
# 3. Missing value check
kg_df.isnull().sum()
```

```python
# Example imputation strategy
kg_df["avg_monthly_income"] = kg_df["avg_monthly_income"].fillna(kg_df["avg_monthly_income"].median())
kg_df.isnull().sum()
```

```python
# 4. Duplicate detection
print("Before removing duplicates:", kg_df.shape)
kg_df = kg_df.drop_duplicates()
print("After removing duplicates:", kg_df.shape)
```

```python
# 5. Type validation
kg_df["year"] = kg_df["year"].astype(int)
kg_df.info()
```

## Why Students Lose Marks Here

A common weak project pattern is:

- they clean the data,
- but do not explain what they cleaned,
- or they perform operations without justification.

A stronger notebook clearly answers:
- What was wrong with the raw data?

# 04.03 Data Transformation and Feature Engineering

After cleaning, students should transform the data when it improves the analysis.

Transformation is not required for its own sake.
It should help:
- comparison,
- interpretation,
- trend analysis,
- or dashboard clarity.

## Useful Transformation Options

- Create ratios
- Compute growth rates
- Build derived indicators
- Encode categories (if needed)
- Normalize or scale (if needed)
- Reshape using pivot / melt
- Merge with another dataset (optional)

```python
# Derived indicators
kg_df["income_per_1000_population"] = kg_df["avg_monthly_income"] / (kg_df["population"] / 1000)
kg_df["employment_rate"] = 100 - kg_df["unemployment_rate"]

kg_df[["region", "year", "avg_monthly_income", "income_per_1000_population", "employment_rate"]].head()
```

```python
# Growth rate example
kg_df = kg_df.sort_values(["region", "year"])
kg_df["income_growth_pct"] = kg_df.groupby("region")["avg_monthly_income"].pct_change() * 100
kg_df.head(10)
```

```python
# Pivot table for trend comparison
pivot_income = kg_df.pivot_table(values="avg_monthly_income", index="year", columns="region")
pivot_income.head()
```

## Transformation Principle

Students should be able to answer:

> “Why did I create this new variable?”

If the answer is unclear, the transformation may not be useful.

# 04.04 Applied Statistics and Exploratory Statistical Analysis

This section is central because the project explicitly requires:

- **Applied Statistics**
- **Exploratory Statistical Analysis**

## Recommended Statistical Components

A strong project should usually include:

### Descriptive Statistics
- mean
- median
- minimum / maximum
- standard deviation
- variance
- quartiles / IQR

### Grouped Statistical Analysis
- region-level summaries
- year-level summaries
- top / bottom comparisons

### Relationship Analysis
- correlation matrix
- Pearson correlation
- Spearman correlation (if appropriate)

### Comparative / Categorical Analysis
- counts
- cross-tabulations
- chi-square (only if relevant and justified)

### Optional Advanced Elements
- simple regression
- trend fitting
- moving averages (for time-related data)

```python
# Descriptive statistics
kg_df.describe()
```

```python
# Grouped statistics by region
region_summary = kg_df.groupby("region")[[
    "population", "unemployment_rate", "avg_monthly_income",
    "school_enrollment_rate", "hospital_beds_per_1000"
]].mean().round(2)

region_summary
```

```python
# Yearly summaries
year_summary = kg_df.groupby("year")[[
    "avg_monthly_income", "unemployment_rate", "school_enrollment_rate"
]].mean().round(2)

year_summary
```

```python
# Correlation matrix
corr = kg_df[[
    "population", "unemployment_rate", "avg_monthly_income",
    "school_enrollment_rate", "hospital_beds_per_1000", "employment_rate"
]].corr()

corr
```

```python
# Correlation heatmap
sns.heatmap(corr, annot=True, fmt=".2f", cmap="rocket_r")
plt.title("Correlation Heatmap")
plt.show()
```

## Statistical Interpretation Guidance

Students should avoid weak interpretation such as:

- “The highest mean is best” (without context)
- “The variables are related because the scatter plot looks upward”
- “Correlation proves cause”

Instead, stronger language is:

- “This variable shows a higher average in these regions.”
- “The heatmap suggests a possible association.”
- “This pattern may indicate a relationship, but not causation.”

# 04.05 Visualization Strategy for the Final Project

The project explicitly requires visualization using multiple plot types discussed in the course.

Students should not use plots mechanically.
Each plot should answer a specific analytical question.

## Recommended Plot Types

- **Bar chart** → compare regions or categories
- **Line chart** → show trends over time
- **Histogram** → show distribution
- **Boxplot** → compare spread across groups
- **Scatter plot** → show relationships
- **Heatmap** → show correlation or matrix summaries
- **Count plot** → show category frequencies

```python
# 1. Bar chart: average income by region
kg_df.groupby("region")["avg_monthly_income"].mean().sort_values().plot(kind="bar")
plt.title("Average Monthly Income by Region")
plt.ylabel("Average Monthly Income")
plt.xticks(rotation=35)
plt.show()
```

```python
# 2. Line chart: average income over time
kg_df.groupby("year")["avg_monthly_income"].mean().plot(marker="o")
plt.title("Average Monthly Income Over Time")
plt.ylabel("Average Monthly Income")
plt.xlabel("Year")
plt.show()
```

```python
# 3. Histogram: unemployment rate
plt.hist(kg_df["unemployment_rate"], bins=10)
plt.title("Distribution of Unemployment Rate")
plt.xlabel("Unemployment Rate")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. Boxplot: school enrollment by region
sns.boxplot(x="region", y="school_enrollment_rate", data=kg_df)
plt.title("School Enrollment Rate by Region")
plt.xticks(rotation=35)
plt.show()
```

```python
# 5. Scatter plot: income vs unemployment
plt.scatter(kg_df["avg_monthly_income"], kg_df["unemployment_rate"])
plt.title("Income vs Unemployment Rate")
plt.xlabel("Average Monthly Income")
plt.ylabel("Unemployment Rate")
plt.show()
```

## Strong Visualization Minimum Standard

A convincing project should usually include at least:

- **1 distribution plot**
- **1 comparison plot**
- **1 time-trend plot**

# 04.06 Dashboarding for the Final Project

This project should now explicitly include **dashboarding**.

Dashboarding is important because it helps transform separate analyses into one coherent communication tool.

## What a Strong Student Dashboard Should Include

### KPI Cards
Examples:
- average monthly income
- average unemployment rate
- average school enrollment
- average hospital beds per 1000
- latest year average income

### Core Charts
Examples:
- regional comparison bar chart
- time trend line chart
- distribution histogram
- relationship scatter plot

### Supporting Table
Examples:
- top 5 regions by income
- bottom 5 regions by school enrollment
- regions needing attention

```python
# KPI summary dictionary
dashboard_kpis = {
    "Average Income": round(kg_df["avg_monthly_income"].mean(), 2),
    "Average Unemployment": round(kg_df["unemployment_rate"].mean(), 2),
    "Average School Enrollment": round(kg_df["school_enrollment_rate"].mean(), 2),
    "Average Hospital Beds/1000": round(kg_df["hospital_beds_per_1000"].mean(), 2),
    "Latest Year Avg Income": round(kg_df[kg_df["year"] == kg_df["year"].max()]["avg_monthly_income"].mean(), 2)
}

dashboard_kpis
```

```python
# Static dashboard using Matplotlib subplots
fig, axes = plt.subplots(2, 2, figsize=(13, 9))

# Regional comparison
kg_df.groupby("region")["avg_monthly_income"].mean().sort_values().plot(kind="bar", ax=axes[0, 0])
axes[0, 0].set_title("Avg Income by Region")
axes[0, 0].tick_params(axis="x", rotation=35)

# Trend over time
kg_df.groupby("year")["avg_monthly_income"].mean().plot(marker="o", ax=axes[0, 1])
axes[0, 1].set_title("Avg Income Over Time")

# Distribution
axes[1, 0].hist(kg_df["unemployment_rate"], bins=10)
axes[1, 0].set_title("Unemployment Rate Distribution")
axes[1, 0].set_xlabel("Unemployment Rate")

# Relationship
axes[1, 1].scatter(kg_df["avg_monthly_income"], kg_df["unemployment_rate"])
axes[1, 1].set_title("Income vs Unemployment")
axes[1, 1].set_xlabel("Avg Income")
axes[1, 1].set_ylabel("Unemployment Rate")

plt.suptitle("Kyrgyzstan Socio-Economic Dashboard (Static)")
plt.tight_layout()
plt.show()
```

```python
# Example supporting table: top 5 regions by average income
kg_df.groupby("region")["avg_monthly_income"].mean().sort_values(ascending=False).head(5).reset_index()
```

## Dashboarding Principle (Very Important)

A dashboard is **not** simply “many charts placed together.”

A strong dashboard should:

- begin with the most important summary values,
- show the most relevant comparisons first,
- support a short spoken explanation,
- and make the final defense easier and more convincing.

## Recommended Minimum Dashboard Standard

For this course, a strong dashboard section should contain:

- **3–5 KPI summaries**
- **3–4 coordinated charts**
- **1 supporting summary table**
- **1 short interpretation paragraph**

# 04.09 Defense Preparation (10–15 Minutes + Q&A)

A technically strong project can still create a weak impression if the defense is unclear.

Students should therefore prepare the defense as seriously as the notebook.

## Recommended 10–15 Minute Defense Flow

### Minute 1–2: Introduction
- What dataset did you choose?
- Why is it relevant to Kyrgyzstan?
- What is the project objective?

### Minute 3–5: Data Wrangling
- What problems existed in the raw data?
- How did you clean them?
- Why did you choose these cleaning strategies?

### Minute 6–9: Applied Statistics + EDA
- What key statistical summaries did you compute?
- Which patterns stood out?
- Which visualizations were most useful?

### Minute 10–12: Dashboard + Key Insights
- Show your dashboard
- Explain 3 major findings
- Explain why the dashboard helps communicate the story

### Minute 13–15: Conclusion
- What did you learn?
- What are the limitations?
- What would you improve in future work?

## Common Q&A Questions

1. Why did you choose this dataset?
2. Why is it suitable for this course project?
3. What were the biggest cleaning challenges?
4. Why did you choose this missing-value strategy?
5. Why did you create these derived variables?
6. Why did you choose these specific plots?
7. Why is your dashboard useful instead of only separate charts?
8. What are your three strongest insights?
9. Did you find correlation or only visual association?
10. What are the limitations of your dataset?
11. If you had more time, what would you improve?

## Strong Answering Strategy

A strong defense answer is usually:

- short,
- evidence-based,
- honest,
- and connected to the data.

Example:

> “Based on this dataset, the pattern suggests a possible relationship, but I would avoid claiming causation without stronger evidence or additional data.”
